# 1. What is Dimensionality Reduction? Why is it important in machine learning?

Answer : Dimensionality reduction is the process of reducing the number of random variables or features under consideration by obtaining a set of principal variables.It transforms data from a high-dimensional space into a low-dimensional space, attempting to retain as much meaningful information (variance) as possible.

It is important in machine learning for several reasons:

1. Curse of Dimensionality: As the number of features increases, the data becomes increasingly sparse, making it difficult for models to learn effectively. This also leads to a significant increase in computational cost and memory requirements.
2. Visualization: It is impossible to visualize data in more than three dimensions. Dimensionality reduction techniques like PCA can project data onto 2D or 3D plots, allowing humans to identify patterns, clusters, and outliers.
3. Noise Reduction: The process often helps in filtering out noise and redundant features, leading to cleaner data and improved model performance.
4. Preventing Overfitting: Models trained on a lower number of relevant features are less likely to overfit to the noise in the training data.

# 2.  Name and briefly describe three common dimensionality reduction techniques.

Answer : 1. Principal Component Analysis (PCA): PCA is a linear technique that identifies the axes (principal components) that maximize the variance in the data. The first component captures the most variance, the second (orthogonal to the first) captures the next most, and so on. It is widely used for data compression and visualization.
2. t-Distributed Stochastic Neighbor Embedding (t-SNE): t-SNE is a non-linear technique particularly well-suited for visualizing high-dimensional data in 2D or 3D. It works by converting similarities between data points to joint probabilities and tries to minimize the divergence between these probabilities in the high-dimensional and low-dimensional spaces, effectively preserving the local structure and revealing clusters.
3. Autoencoders: Autoencoders are a type of neural network used for unsupervised learning. They are designed to learn a compressed, dense representation (encoding) of the input data by training the network to ignore signal "noise." The network is trained to reconstruct its inputs, and the bottleneck layer in the middle serves as the dimensionality-reduced representation.


# 3. What is clustering in unsupervised learning? Mention three popular clustering algorithms.

Answer:
Clustering is an unsupervised learning technique that involves grouping a set of objects in such a way that objects in the same group(called a cluster) are more similar to each other than to those in other groups. The goal is to discover the inherent, natural groupings within the unlabeled data.

Three popular clustering algorithms are:

1. K-Means Clustering: A centroid-based algorithm that partitions the data into K pre-defined, non-overlapping clusters. Each data point belongs to the cluster with the nearest mean (centroid).
2. Hierarchical Clustering: A connectivity-based algorithm that builds a multilevel hierarchy of clusters by either a "bottom-up" (agglomerative) approach, where each point starts as its own cluster and pairs are merged, or a "top-down" (divisive) approach.
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise): A density-based algorithm that groups together closely packed points, marking points that lie alone in low-density regions as outliers. It does not require specifying the number of clusters beforehand.

# 4. Explain the concept of anomaly detection and its significance.

Answer:
Anomaly detection,also known as outlier detection, is the process of identifying rare items, events, or observations that raise suspicions by differing significantly from the majority of the data. These anomalies often translate to critical, actionable information in various domains.

Its significance is immense:

· Fraud Detection: Identifying fraudulent credit card transactions or insurance claims.
· Network Security: Detecting intruders or malicious activity in network traffic.
· Fault Detection: Monitoring industrial machinery to identify defects or failures before they cause significant damage.
· Healthcare: Spotting unusual patterns in medical images or patient vital signs that may indicate disease.


# 5. List and briefly describe three types of anomaly detection techniques.

Answer:

1. Statistical Methods: These methods assume that normal data points follow a statistical model (e.g., a Gaussian distribution) and flag points that have a low probability of being generated by that model as anomalies. Examples include Z-score and Grubbs' Test.
2. Proximity-Based Methods: These methods are based on the idea that normal data points exist in dense neighborhoods, while anomalies are far from their nearest neighbors.
   · K-Nearest Neighbors (KNN): Anomalies are points whose average distance to their k-nearest neighbors is large.
3. Model-Based Methods: These techniques use machine learning models to learn the pattern of normal data. Any data point that does not fit this model well is considered an anomaly.
   · Isolation Forest: This method explicitly isolates anomalies by randomly selecting a feature and then a split value between the max and min of that feature. Anomalies are easier to isolate, so they require fewer random splits.
   · Autoencoders: An autoencoder trained on normal data will learn to reconstruct it well. When presented with an anomaly, the reconstruction error will be high, signaling an outlier.


# 6. What is time series analysis? Mention two key components of time series data.

Answer:
Time series analysis involves analyzing time-ordered data points collected at specific intervals to extract meaningful statistics,characteristics, and patterns. The goal is to understand the underlying structure of the data, which can be used for forecasting future values.

Two key components of time series data are:

1. Trend: The long-term increase or decrease in the data over a significant period. It represents the underlying direction of the data (e.g., a consistent upward trend in company revenue over five years).
2. Seasonality: The presence of variations that occur at specific, regular intervals due to seasonal factors. These patterns repeat over a known, fixed period (e.g., daily, weekly, or yearly), such as increased ice cream sales every summer.

# 8. Write Python code to perform K-means clustering on a sample dataset.(include your Python code and output in the code box below.)

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import os
import warnings

# Suppress the specific warnings
warnings.filterwarnings('ignore', category=UserWarning)

# 1. Create a sample dataset
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# 2. Visualize the initial data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Sample Dataset (Before Clustering)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# 3. Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# 4. Visualize the clustered data
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()

# Print the cluster centers
print("Cluster Centers (Centroids):")
print(centers)
```

# 9. What is inheritance in OOP? Provide a simple example in Python.

Answer:
Inheritance is a fundamental concept in Object-Oriented Programming(OOP) that allows a new class (called a child or derived class) to inherit attributes and methods from an existing class (called a parent or base class). This promotes code reusability and establishes a natural hierarchy between classes.

```python
# Parent class
class Animal:
    def speak(self):
        return "This animal makes a sound"

# Child class (inherits from Animal)
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Using the classes
a = Animal()
d = Dog()

print(a.speak())   # Output: This animal makes a sound
print(d.speak())   # Output: Dog barks
```

# 10. How can time series analysis be used for anomaly detection?

Answer:
Time series analysis is highly effective for anomaly detection because it models the normal behavior of a system over time,including its trend, seasonality, and cyclic patterns. Any significant deviation from this learned normal pattern can be flagged as an anomaly.

The process typically involves these steps:

1. Model the Time Series: A model (e.g., ARIMA, Exponential Smoothing, or a modern deep learning model like LSTM) is trained on historical data to learn and capture the underlying patterns (trend, seasonality).
2. Forecast Future Values: The trained model is used to predict the next value(s) in the series, along with a prediction interval (confidence band).
3. Calculate Residuals: The difference between the actual observed value and the model's forecasted value is calculated (this is called the residual or error).
4. Flag Anomalies: Anomalies are detected based on these residuals. A data point can be flagged as an anomaly if:
   · Its residual is significantly larger than the historical residuals.
   · The actual value falls outside the forecasted prediction interval (e.g., outside the 95% confidence band).

This approach is widely used to detect unusual spikes in website traffic, sudden drops in stock prices, or unexpected sensor readings in industrial equipment.
